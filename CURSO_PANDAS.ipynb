##### Conditional Selection

```python
# Sometimes we need to ask questions based on conditions

NBA.Tm == 'POR'
```

```python
# We can use the pipe (|) to bring a question or other question:

NBA.loc[(NBA.Tm == 'POR') | (NBA.PTS >= 7)]
```

```python
# isin() -> to select data whose value 'is in' a list of values
# it bring me a values specify from a list of values

NBA.loc[NBA.Tm.isin(['POR', 'ATL'])]
```

```python
# Bring the values that is not null

NBA.loc[NBA.PTS.notnull()]
```

```python
#DATASET LOAD
reviews = pd.read_csv('winemag-data-130k-v2.csv', sep = ',', index_col = 0)
reviews.head()
```

##### Sumary Functions

```python
# This method generates a high-level summary of the attributes of the given column
# The output bellow only makes sense for numerical data.
reviews.points.describe()
```

```python
# for string:
reviews.taster_name.describe()
```

```python
# To get some particular simple sumarry statisstic about a column in a DataFrame or a Series.

reviews.points.mean()
```

```python
# To see uniques values:

reviews.taster_name.unique()
```

```python
# To see a list of unique values and how often they occur in the dataset, we can use the 'value_counts()'

reviews.taster_name.value_counts()
```

##### Maps

```python
# Map: used for substituting each value in a Series with another value
reviews_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - reviews_points_mean)
```

##### Apply

```python
def remean_points(row):
        row.points = row.points - reviews_points_mean
        return row

reviews.apply(remean_points, axis = 'columns')
```

```python
reviews_points_mean = reviews.points.mean()
reviews.points - reviews_points_mean
```

```python
reviews.country + ' - ' + reviews.region_1
```

# Class 04

##### Group By

```python
# Often we want to group our data, and then do something specific to the group the data is in

reviews.groupby('points').points.count()
```

```python
# To get the cheapest wine in each point value category, we can do the following:
reviews.groupby('points').price.min()
```

```python
reviews.groupby('winery').winery.count()
```

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# We can also group by more than one column:

reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# Another group by method is 'agg()', with this function we can run a bunch of different functions on the DataFrame simultaneously.

reviews.groupby('country').price.agg([len, max, min])
```

##### Multi-Indexes

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
ind = countries_reviewed.index
type(ind)
```

```python
# In general the multi-index method you will use most often is the one for converting back to a regular index, the 'reset_index()' method:
countries_reviewed.reset_index()
```

##### Sorting

```python
# To get data in the order

countries_reviewed =  countries_reviewed.reset_index()
countries_reviewed.sort_values(by = 'len')
```

```python
# descending sort

countries_reviewed.sort_values(by = 'len', ascending = False)
```

```python
# to sort by index values

countries_reviewed.sort_index()
```

```python
# we can sort more than once column at a time

countries_reviewed.sort_values(by = ['country', 'len'])
```

```python
reviews.groupby('price').points.agg('max').sort_index(ascending = False)
```

##### Combining

```python
canadian_youtube = pd.read_csv('CAvideos.csv', sep = ',')
british_youtube = pd.read_csv('GBvideos.csv', sep = ',')
```

```python
# 40881 rows
canadian_youtube
```

```python
# 38915 rows
british_youtube
```

```python
# concat -> This is useful when we have data in different DataFrame or Series objects but having the same fields(columns)

# 79797 rows
pd.concat([canadian_youtube, british_youtube])
```

```python
# join()

left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix = '_CAN', rsuffix = '_UK')
```
