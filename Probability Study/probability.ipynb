# Probability Study with Python

```python
import pandas as pd
import numpy as np
```

```python
math = pd.read_csv('student-mat.csv', sep = ',')
math.head()
```

```python
len(math)
```

```python
math['Grade_A'] = np.where(math['G3'] * 5 >= 80, 1, 0)
math.head()
```

```python
math['high_absenses'] = np.where(math['absences'] >= 10, 1, 0)
math.head()
```

```python
math['count'] = 1
math.head()
```

```python
math = math[['Grade_A', 'high_absenses', 'count']]
math.head()
```

```python
pd.pivot_table(
    math,
    values = 'count',
    index = ['Grade_A'],
    columns = ['high_absenses'],
    aggfunc = np.size,
    fill_value = 0
)
```

### Now, let's calculate the probability a student gets an A (80%+) in math, given they miss 10 or more classes.

#### P(A) is the probability of a grade of 80% or greater

#### P(B) is the probability of missing 10 or more classes

#### P(A|B) is the probability of a 80%+ grade, given missing 10 or more classes

```python
A = ((35+5)/(35+5+277+78)) * 100
print(f'The probability of a grade of 80% or greater is {A:.2f} %')
```

```python
B = ((78+5)/(78+5+35+277)) * 100
print(f'The probability of missing 10 or more classes is {B:.2f} %')
```

```python
intersec = ((5)/(277+5+78+35)) * 100
print(f'The probability of a 80%+ grade, given missing 10 or more classes is {intersec:.2f} %')
```

#### The conditional probability formula is P(A|B) = P(Aâˆ©B)/P(B)

```python
conditional = (intersec/B) * 100
print(f'The conditional probability is {conditional:.2f} %')
```
