## Class 2

#### Line Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_df = pd.read_csv('spotify.csv', sep = ',', index_col = 'Date', parse_dates = True)
```

```python
spotify_df.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize = (14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song
sns.lineplot(data = spotify_df)
```

#### Plot a subset of the data

```python
spotify_df.columns.unique()
```

```python
# Set the width and height of the figure
plt.figure(figsize = (14,6))

# Add title
plt.title("Daily GLobal Streams of Popular Song in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data = spotify_df['Shape of You'], label = 'Shape of You')

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data = spotify_df['Despacito'], label = 'Despacito')

# Add label for horizontal axis
plt.xlabel('Date')
```

# Class 3

#### Bar Charts and Heatmaps

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flight = pd.read_csv('flight_delays.csv', sep = ',', index_col = 'Month')
flight
```

```python
# Bar Chart

#Set the width and height of the figure
plt.figure(figsize = (10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x = flight.index, y = flight['NK'])

# Add label for vertical axis
plt.ylabel("Arrival Delay (in Minutes) ")
```

```python
# Heatmap

plt.figure(figsize = (14,7))


plt.title("Average Arrival Delay for Each Airline, by Month")


sns.heatmap(data = flight, annot = True)

plt.xlabel("Airline")
```

# Class 4

#### Scatter Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
insurance_data = pd.read_csv("insurance.csv", sep = ',')
insurance_data.head()
```

```python
# to create a scatter plot , we use the 'sns.scatterplot' command and specify the values for:
# the horizontal x-axis (x = insurance_data['bmi']), and
# the vertical y-axis (y = insurance_data['charges']).

sns.scatterplot(x = insurance_data['bmi'], y = insurance_data['charges'])
```

```python
# regression line

sns.regplot(x = insurance_data['bmi'], y = insurance_data['charges'])
```

##### Color-coded scatter plots

```python
# We can use scatter plots to display the relationships between three variables.

sns.scatterplot(x = insurance_data['bmi'], y = insurance_data['charges'], hue = insurance_data['smoker'])
```

```python
# Two regression lines (smokers and nonsmokers)

sns.lmplot(x = 'bmi', y = 'charges', hue = 'smoker', data = insurance_data)

```

```python
sns.swarmplot (x = insurance_data['smoker'], y = insurance_data['charges'])
```
